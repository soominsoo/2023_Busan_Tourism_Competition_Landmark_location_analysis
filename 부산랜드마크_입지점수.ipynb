```python
import pandas as pd
import numpy as np
```

```python
전처리데이터셋 = '입지분석_점수.csv'
pd.read_csv('입지분석_점수.csv',encoding='cp949')
```

```python
df = pd.read_csv(f'{전처리데이터셋}', header=0, encoding='cp949', engine='python')
```

```python
df.isna().sum()
```

```python
df.fillna(0,inplace=True)
```

```python
df['교통접근성']=(df['버스정류장']*0.4)+(df['지하철역']*0.6)
```

```python
df.drop(['버스정류장','지하철역'],axis=1,inplace=True)
```

```python
점수칼럼=['인구수', '교통접근성', '맛집', '해안선','산면적','문화재', '명소', '중심관광지']
자연_점수칼럼 = ['인구수', '교통접근성', '맛집', '해안선','산면적']
관광_점수칼럼 = ['인구수', '교통접근성', '맛집','문화재', '명소', '중심관광지']
```

```python
자연_가중치 = [1,2,1,4,3]
관광_가중치=[1,2,1,3,3,3]
```

```python
자연=['해안선','산면적']
공통=['인구수', '교통접근성', '맛집']
자연_가=[4,3]
공통=[1,2,1]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[점수칼럼] = scaler.fit_transform(df.loc[:,점수칼럼])
```

```python
df_자연=df.drop(['문화재', '명소', '중심관광지'],axis=1)
df_관광=df.drop(['산면적','해안선'],axis=1)
```

```python
for i in range(0, len(관광_점수칼럼)):
    name = 관광_점수칼럼[i]
    df_관광[name] = df_관광[name] * 관광_가중치[i] 

df_관광['sum'] = df_관광[관광_점수칼럼].sum(1)
df_관광['순위'] = df_관광['sum'].rank(method='dense', ascending=False).astype(int)
df_관광 = df_관광.sort_values(by='순위')
df_관광
```

```python
for i in range(0, len(자연)):
    name = 자연[i]
    df_자연[name] = df_자연[name] * 자연_가[i] 
    
df_자연['sum'] = df_자연[자연].sum(1)
df_자연['순위'] = df_자연['sum'].rank(method='dense', ascending=False).astype(int)
df_자연 = df_자연.sort_values(by='순위')
df_자연
```

```python
df_자연.to_csv('C:/Users/soo/!busan/project/정제파일/자연_점수1.csv', index=False, sep=',', encoding='cp949')
df_관광.to_csv('C:/Users/soo/!busan/project/정제파일/관광_점수.csv' ,index=False, sep=',', encoding='cp949')
```

```python
df = pd.read_csv(f'{전처리데이터셋}', header=0, encoding='cp949', engine='python')
```

```python
df.drop(['sum','순위'],axis=1,inplace=True)
```

```python
공통=['인구수', '교통접근성', '맛집','산면적','해안선']
공통_가중치=[1,3,1,2,2]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[공통] = scaler.fit_transform(df.loc[:,공통])
```

```python
df.describe()
```

```python
for i in range(0, len(공통)):
    name = 공통[i]
    df[name] = df[name] * 공통_가중치[i] 

df['sum'] = df[공통].sum(1)
df['순위'] = df['sum'].rank(method='dense', ascending=False).astype(int)
df = df.sort_values(by='순위')
df
```
